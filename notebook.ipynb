```python
# Predict heart disease based off of physical characteristics

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import csv
from sklearn.metrics import mean_absolute_error


# # Our inputs
# age = input("Input your current age: ")
# sex = input("Input your gender (1 female, 0 male): ")
# cp = input("Input chest pain (0 - 3): ")
# trestbps = input("Input resting bloog pressure (about 90 - 200): ")
# chol = input("Input your serum cholesterol (120 - 570): ")

# # Headers and input data
# header = ["age", "sex", "trestbps", "chol"]
# data = [age, sex, trestbps, chol]

# # User input
# with open("test.csv", "w") as file:
#     # CSV WRITER
#     writer = csv.writer(file)
    
#     # Write headers and data
#     writer.writerow(header)
#     writer.writerow(data)


dataset = pd.read_csv("./heart.csv")
test_dataset = pd.read_csv("./test_2.csv")


# Test truth
test_truth = test_dataset["target"]

# Function to convert string target to number
def convertTarget(target):
    if target == "Presence":
        return 1
    else:
        return 0

# Map over all targets and turn them into numbers
test_truth = list(map(convertTarget, test_truth))


# Training target
train_y = dataset["target"]

# Training data
# Our category variables
data_variables = ["age", "sex", "trestbps", "chol"]
data = dataset[data_variables]


# Format categorical data into numerical
data = pd.get_dummies(data)


# Grab only the categorical data variables
test_data = test_dataset[data_variables]

# Turn categorical data into numerical
test_data = pd.get_dummies(test_data)



# Split the data into test and train
x_train, x_test, y_train, y_test = model_selection.train_test_split(data, train_y)

model = RandomForestClassifier(n_estimators=100)

model.fit(x_train, y_train)

predict = model.predict(test_data)

print("MAE: {}".format(mean_absolute_error(test_truth, predict)))

for prediction in predict:
    if prediction == 0:
        print("Present")
    else:
        print("Absent")
```
